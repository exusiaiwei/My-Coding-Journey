```r
required_packages <- c("tidyverse", "ggplot2","paletteer")

install_and_load <- function(package) {
  if (!require(package, character.only = TRUE)) {
    install.packages(package, dependencies = TRUE)
    library(package, character.only = TRUE)
  }
}

invisible(lapply(required_packages, install_and_load))
```

```r
csv_file <- "output_files/Multiple Text Project 1 - Results.csv"
data <- read_csv(csv_file)
```

```r
type_plot <- data %>%
  gather(key = "Metric", value = "Value", TTR, `h-Point`, Entropy, R1) %>%
  ggplot(aes(x = Type, y = Value, color = Type)) +
  geom_jitter(alpha = 0.7, size = 2) +  # 散点图
  facet_wrap(~ Metric, scales = "free_y") +
  theme_minimal() +
  scale_color_paletteer_d("tvthemes::BluebirdAzurite") +
  labs(title = "Type 对各指标的分布", x = "Type", y = "Value")

print(type_plot)
```

```r
model_plot <- data %>%
  gather(key = "Metric", value = "Value", TTR, `h-Point`, Entropy, R1) %>%
  ggplot(aes(x = Model, y = Value, color = Model)) +
  geom_jitter(alpha = 0.7, size = 2) +  # 散点图
  facet_wrap(~ Metric, scales = "free_y") +
  theme_minimal() +
  scale_color_paletteer_d("tvthemes::BluebirdAzurite") +
  labs(title = "Model 对各指标的分布", x = "Model", y = "Value")
print(model_plot)
```

```r
ttr_plot <- data %>%
  ggplot(aes(x = Model, y = TTR, fill = Type, alpha = Type)) +
  geom_boxplot() +
  scale_fill_paletteer_d("unikn::pal_seegruen") +
  scale_alpha_manual(values = rep(c(1, 0.5), length.out = length(unique(data$Type)))) +
  theme_minimal() +
  theme(legend.position = "none") +
  labs(title = "TTR 的箱线图", x = "", y = "TTR")

# 打印 TTR 箱线图
print(ttr_plot)
```
